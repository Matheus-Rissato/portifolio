# Aula 02 - Importação, padrozinação de dados e criação do dataset usando o XLS

```python
import pandas as pd
```

```python
df_principal = pd.read_excel("/content/acoes_puras.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/acoes_puras.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/acoes_puras.xlsx", sheet_name="Ticker")
```

```python
df_chatgpt = pd.read_excel("/content/acoes_puras.xlsx", sheet_name="Chat GPT - Segmento e idade")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data', 'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'})
df_principal
```

```python
df_principal['var_dia_pct_real'] = df_principal['var_dia_pct']/100
df_principal
```

```python
df_principal['var_inicial'] = df_principal['valor_final']/(df_principal['var_dia_pct_real'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'})
df_principal['varicao_reais'] = (df_principal['valor_final'] - df_principal['var_inicial'])*df_principal['qtd_teorica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['qtd_teorica'] = df_principal['qtd_teorica'].astype(int)
df_principal
```

```python
df_principal['Resultado'] = df_principal['varicao_reais'].apply(lambda x: "Subiu" if x>0 else ("Desceu" if x < 0  else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome')
df_principal = df_principal.rename(columns={'Nome':'nome_da_empresa'})
df_principal = df_principal.rename(columns={'Segmento':'segmento', 'Idade em anos':'idade_anos' })
```

```python
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade_anos'].apply(lambda x: "Entre 50 e 100 anos" if x > 50 and x < 100 else ("Mais de 100 anos" if x > 100 else "Menos de 50 anos" ))
df_principal
```

# Aula 03 - Análises

```python
maior_var_reais = df_principal['varicao_reais'].max()
menor_var_reais = df_principal['varicao_reais'].min()
media_var_reais = df_principal['varicao_reais'].mean()
media_var_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['varicao_reais'].mean()
media_var_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['varicao_reais'].mean()

print(f'A maior variação do dia  foi de {"R$ {:.2f}".format(maior_var_reais)}')
print(f'A menor variação do dia  foi de {"R$ {:.2f}".format(menor_var_reais)}')
print(f'A variação média do dia  foi de {"R$ {:.2f}".format(media_var_reais)}')
print(f'A variação média das ações que subiram no dia foi de {"R$ {:.2f}".format(media_var_subiu)}')
print(f'A variação média das ações que desceram no dia foi de {"R$ {:.2f}".format(media_var_desceu)}')
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal.groupby('segmento')['varicao_reais'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['varicao_reais'].sum().reset_index()
texto_grafico = df_analise_saldo['varicao_reais'].apply(lambda x: 'R${:,.2f}'.format(x))
print(df_analise_saldo)
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='varicao_reais', text= texto_grafico, title='Variação Reais por Resultado')
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
dados
```

```python
dados.columns = ['abertura', 'max', 'min', 'fechamento', 'fechamento_ajustado','volume']
dados = dados.rename_axis('data')
dados
```

```python
dados['fechamento'].plot(figsize=(10,6))
plt.title('Variação do preço por data', fontsize=16)
plt.legend(['Fechamento'])
```

```python
df = dados.head(60).copy()

df['Data'] = df.index

df['Data'] = df['Data'].apply(mdates.date2num)
df
```

```python
fig, ax = plt.subplots(figsize=(15,8))

width = 0.7

for i in range(len(df)):
  if df['fechamento'].iloc[i] > df['abertura'].iloc[i]:
    color = 'green'
  else:
    color ='red'

  ax.plot([df['Data'].iloc[i],df['Data'].iloc[i]],
        [df['min'].iloc[i],df['max'].iloc[i]],
        color=color,
        linewidth=1)
  ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['abertura'].iloc[i], df['fechamento'].iloc[i])),
             width,
             abs(df['fechamento'].iloc[i] - df['abertura'].iloc[i]),
             facecolor=color))

df['MA7'] = df['fechamento'].rolling(window=7).mean()
df['MA14'] = df['fechamento'].rolling(window=14).mean()

ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 dias')
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 dias')
ax.legend()

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

plt.grid(True)

plt.show()

```

# Aula 5 - Previsão de Séries Temporais de Ações

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
dados_aula_5 = yf.download('JNJ', start='2020-01-01', end='2023-12-31', progress=False)
dados_aula_5 = dados_aula_5.reset_index()
```

```python
dados_treino = dados_aula_5[dados_aula_5['Date'] < '2023-07-31']
dados_teste = dados_aula_5[dados_aula_5['Date'] >= '2023-07-31']
```

```python
dados_prophet_treino = dados_treino[['Date', 'Close']].rename(columns={'Date':'ds', 'Close':'y'})
```

```python
modelo = Prophet(weekly_seasonality=True,
                 yearly_seasonality=True,
                 daily_seasonality=False)

modelo.add_country_holidays(country_name='US')

modelo.fit(dados_prophet_treino)
```

```python
futuro = modelo.make_future_dataframe(periods=150)
previsao = modelo.predict(futuro)
```

```python
plt.figure(figsize=(14,8))
plt.plot(dados_treino['Date'], dados_treino['Close'], label='Dados de treino', color='blue')
plt.plot(dados_teste['Date'], dados_teste['Close'], label='Dados Reais (teste)', color='green')
plt.plot(previsao['ds'], previsao['yhat'], label='Previsão', color='orange', linestyle='--')

plt.axvline(dados_treino['Date'].max(), color='red', linestyle='--', label='Início da previsão')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Previsão de Preço e Fechamento vs Dados Reais')
plt.legend()
plt.show()
```
